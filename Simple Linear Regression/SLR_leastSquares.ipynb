**Simple Linear Regression using Linear Least Squares Method**

Another method to implement linear regression is 'Linear Least Squares Method'. It is more simple and more interpretable compared to gradient descent algorithm. However it is not a right approach when there's a large dataset having a linear relationship as it only works well for linear models. Here's how it works:

Linear Least Squares is a closed-form solution typically used to solve linear regression problems. The objective is to find the best-fitting line that minimizes the sum of squared residuals which is the difference between the observed and the predicted value.

Suppose there's a dataset given by the equation Y = $\beta_0$ + $\beta_1$X + $ϵ$, where $ϵ$ is the noise, $\beta_0$ & $\beta_1$ are the parameters (intercept and slope) and X is independent variable. (It is assumed that $ϵ$ is generated from a normal distribution with mean zero)

Let the regression coefficient estimates be $\hat{\beta_0}$ & $\hat{\beta_1}$. Then the estimated linear fit line is $\hat{y}$ = $\hat{\beta_0}$ + $\hat{\beta_1}$ $x_i$ where X = $x_i$ for the i-th value.

The residual sum of squares (RSS) is then:

\begin{equation}
RSS = e_1^2 + e_2^2 + e_3^2 + ... + e_n^2
\end{equation}


\begin{equation}
RSS = \sum_{i=1}^{n} e_i^2
\end{equation}

where $e_i$ is the error term.

\begin{equation}
RSS = (Y_1 - (\hat{\beta_0} + \hat{\beta_1}x_1))^2 + (Y_2 - (\hat{\beta_0} + \hat{\beta_1}x_2))^2 + ... + (Y_n - (\hat{\beta_0} + \hat{\beta_1}x_n))^2
\end{equation}

\begin{equation}
RSS = \sum_{i=1}^{n} (Y_i - (\hat{\beta_0} + \hat{\beta_1}x_n))^2
\end{equation}

Using a lot of Calculus (which I will skip here), we can derive the equation for estimates:

\begin{equation}
\hat{\beta_1} = \frac{\sum_{i=1}^{n} (x_i - \bar{x}) (y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
\end{equation}

\begin{equation}
\hat{\beta_0} = \bar{y} - \hat{\beta_1}\bar{x}
\end{equation}

where $\bar{x}$ & $\bar{y}$ are defined as:

\begin{equation}
\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i
\end{equation}

\begin{equation}
\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i
\end{equation}


Now, to finally apply the linear least squares method in Python, first import all the necessaery libraries, and define the number of data points, error term(s) and the data points

```python
import numpy as np
import matplotlib.pyplot as plt

#Generate synthetic data
n = 100  # Number of data points
error = np.random.normal(0, 1, n)  # 100 random noise values
x = np.random.rand(n) * 10  # Random values for x between 0 and 10
y = 2 * x + 1 + error  # y = 2x + 1 with some random noise
```

Now, we calculate the necessary terms and plot the least square line

```python
#Calculate the mean of x & y
x_mean = np.mean(x)
y_mean = np.mean(y)

#Calculate the estimates
beta_1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)
beta_0 = y_mean - beta_1 * x_mean

#Calculate the predicted y values (y_pred) using the regression line
y_pred = beta_1 * x + beta_0
```

```python
#Plotting the graph
plt.scatter(x, y, label='Data points', color='blue')
plt.plot(x, y_pred, label=f'Best fit line: y = {beta_1:.2f}x + {beta_0:.2f}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Method')
plt.legend()
plt.grid(True)
plt.show()


# Output the values of the slope and intercept
print(f"Slope (beta_1): {beta_1:.2f}")
print(f"Intercept (beta_0): {beta_0:.2f}")
```

Since the scatter plot of x and y can be a bit jumbled (depending on how x is generated), it might be helpful to sort x before plotting the best-fit line so that the line is drawn in increasing order of x. You can sort the x values and then compute y_pred accordingly, ensuring the line is smooth, given as:

```python
sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]
y_pred_sorted = beta_1 * x_sorted + beta_0
plt.scatter(x, y, label='Data points', color='blue')
plt.plot(x_sorted, y_pred_sorted, label=f'Best fit line: y = {beta_1:.2f}x + {beta_0:.2f}', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Method')
plt.legend()
plt.grid(True)
plt.show()
```
